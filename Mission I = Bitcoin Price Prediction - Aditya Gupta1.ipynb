### Micro Task 5 : What is the MSE of your model?

```python
# Model Formulation and Prediction
X = bit1.drop('btc_market_price',axis=1)
Y = bit1['btc_market_price']
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 0)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
# Model Building
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()
lreg.fit(X_train, Y_train)
# Training & Testing Score
print('Training Score: ', round(lreg.score(X_train, Y_train)*100,2),'%')
print('Testing Score: ', round(lreg.score(X_test, Y_test)*100,2))
```

```python
# Prediction
y_predict = lreg.predict(X_test)
y_predict
```

```python
from sklearn import metrics
```

```python
acc = metrics.r2_score(Y_test,y_predict)
print('Accuracy Score of Model: ', round(acc*100,2),'%')
```

```python
# Actual v/s Predicted Price
price_prediction = pd.DataFrame({'Actual Price': Y_test, 'Predicted_Price': y_predict})
price_prediction
```

```python
# Model Evaluation
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, y_predict))
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, y_predict))
```

```python
# Rounding off
print('Mean Absolute Error:', round(metrics.mean_absolute_error(Y_test, y_predict),2))
print('Mean Squared Error:', round(metrics.mean_squared_error(Y_test, y_predict),2))
```

### The MSE of my model is 439.06
