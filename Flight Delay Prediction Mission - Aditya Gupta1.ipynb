```python
df_extracted = df_extract.drop(columns = ['YEAR', 'AIRLINE', 'FLIGHT_NUMBER','DISTANCE','TAIL_NUMBER', 
                               'TAXI_OUT','DEPARTURE_TIME', 'SCHEDULED_TIME','WHEELS_OFF', 
                               'ELAPSED_TIME', 'AIR_TIME','WHEELS_ON', 'DAY_OF_WEEK','TAXI_IN', 
                               'ARRIVAL_TIME','CANCELLATION_REASON'], axis = 1)
df_extracted.head()
```

```python
df_extracted.columns
```

```python
column_means = df_extracted.mean()
print(column_means)
df_extracted = df_extracted.fillna(column_means)
```

```python
df_extracted.head()
```

```python
df_extracted['Result'] = 0
df_extracted.loc[df_extracted['ARRIVAL_DELAY']>15, 'Result'] = 1
df_extracted.head()
```

```python
df_extracted['Result'].value_counts()
```

```python
df_extracted.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
x = df_extracted[['MONTH','DAY', 'SCHEDULED_DEPARTURE', 'DEPARTURE_DELAY', 'SCHEDULED_ARRIVAL', 'DIVERTED', 'CANCELLED',
                  'AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']]
y = df_extracted['Result']

scaler = StandardScaler()

X_train, X_test, y_train, y_test = train_test_split(x , y, test_size = 0.30, random_state = 42)
X_train = scaler.fit_transform(X_train)

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
X_test = scaler.transform(X_test)
yhat = clf.predict(X_test)
print(roc_auc_score(yhat, y_test))
```
